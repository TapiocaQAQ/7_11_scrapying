```python
import urllib.request
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time as t
import requests
import numpy as np
from PIL import Image

if __name__ == "__main__":
    # account & password
    PHONE = 'enter Ur phone'
    PASSWORD = 'enter Ur password'
    
    options = Options()
#     options.add_argument('--headless')
#     options.add_argument('--no-sandbox')
    driver = webdriver.Chrome('C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe', chrome_options=options)
    #driver.set_window_size(1280, 1280)
    
    
    driver.get("https://myship.7-11.com.tw/Home/Main")
    driver.find_element_by_xpath('//*[@class="header-small-btn "]').click()
    test = driver.find_element_by_xpath('//*[@class="header-small-btn "]')
    print(test)
    print("Start : %s" % t.ctime())
    t.sleep(5)
    print("end : %s" % t.ctime())
    driver.find_element_by_xpath('//button[@class="btnOPLogin btn-login btn-login-op"]').click()
    
    print("Start : %s" % t.ctime())
    t.sleep(3)
    print("end : %s" % t.ctime())
    
    img_uri = str(driver.find_element_by_xpath('//img[@id="imgVerify"]').get_attribute('src'))
    
    response = urllib.request.urlopen(img_uri)
    
    with open('image.png', 'wb') as f:
        f.write(response.file.read())
        f.close()
        
    img = Image.open('image.png')
    img.show()
    
    val = input("Enter your value: ")
    
    driver.find_element_by_xpath('//input[@name="verifyCode"]').send_keys(str(val))
    driver.find_element_by_xpath('//input[@id="mobile"]').send_keys(PHONE)
    driver.find_element_by_xpath('//input[@id="passwd_e"]').send_keys(PASSWORD)
    driver.find_element_by_xpath('//a[@id="loginBtn"]').click()
    
    print("Start : %s" % t.ctime())
    t.sleep(3)
    print("end : %s" % t.ctime())
    
    driver.get("https://myship.7-11.com.tw/seller/order/All")
    soup = bs(driver.page_source,"html.parser")
    print(soup)
```
